```python
import os
import shutil

def copy_large_files(source_folder, target_folder, check_folder, size_limit_kb = 640):
    os.makedirs(target_folder, exist_ok = True)  # Ensure target folder exists

    for filename in os.listdir(source_folder):
        source_path = os.path.join(source_folder, filename)
        check_path = os.path.join(check_folder, filename)
        target_path = os.path.join(target_folder, filename)

        # Ensure the file exists in Folder C before checking its size
        if os.path.exists(check_path):
            file_size_kb = os.path.getsize(check_path) / 1024  # Convert bytes to KB
            
            if file_size_kb < size_limit_kb:
                print(f"Processing: {filename}")
                shutil.copy2(source_path, target_path)
                print(f"Copied: {filename}")
            else:
                print(f"Skipped: {filename} (File in Folder C is {file_size_kb:.2f} KB)")
        else:
            print(f"Skipped: {filename} (File not found in Folder C)")

# Example usage
folder_a = "/Volumes/arxivBeta/_Tobias/Opera/20250204/Importer"
folder_b = "/Volumes/arxivBeta/_Tobias/Opera/20250204/Importer_cases"
folder_c = "/Volumes/arxivBeta/_Tobias/Opera/20250204/Montages_cases_V2"

copy_large_files(folder_a, folder_b, folder_c)
```
