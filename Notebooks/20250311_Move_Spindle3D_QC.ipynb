```python
import os
from pathlib import Path
import shutil
```

```python
QC_path = '/Volumes/arxivBeta/_Tobias/Opera/20250314/05_Spindle3D_QC_Juptyer'
Spindle3D_path = "/Volumes/arxivBeta/_Tobias/Opera/20250314/04_Spindle3D"
```

```python
def organize_subfolders(folder_A, folder_B):
    # Define source and destination root directories
    folder_a_root = Path(folder_A)
    folder_b_root = Path(folder_B)
    
    # Dictionary to store PNG filenames found in A/A1 and A/A2
    png_files = {'Bipolar': [], 'Other': []}
    
    # Scan both subdirectories in folder A for PNG files
    for subdir in ['Bipolar', 'Other']:
        folder_path = folder_a_root / subdir
        if folder_path.exists():
            # Get base filenames without extensions
            png_files[subdir] = [file.stem.replace('-out', '') for file in folder_path.glob('*.png')]
    
    # Get all subfolders in folder B
    b_subfolders = [folder for folder in folder_b_root.iterdir() if folder.is_dir()]
    
    # Create destination directories if they don't exist
    for subdir in ['Bipolar', 'Other']:
        dest_dir = folder_b_root / subdir
        dest_dir.mkdir(parents = True, exist_ok = True)
    
    # Move matching subfolders based on PNG files
    for src_folder in b_subfolders:
        folder_name = src_folder.name
        for subdir, png_list in png_files.items():
            if folder_name in png_list:
                dest_path = folder_b_root / subdir / folder_name
                
                # Check if destination already exists
                if dest_path.exists():
                    print(f"Warning: Destination {dest_path} already exists. Skipping {src_folder}")
                    continue
                    
                try:
                    shutil.move(str(src_folder), str(dest_path))
                    print(f"Moved {src_folder.name} to {dest_path}")
                except Exception as e:
                    print(f"Error moving {src_folder}: {str(e)}")

if __name__ == "__main__":
    organize_subfolders(folder_A = QC_path, folder_B = Spindle3D_path)
```
